# Limpeza dos Dados

Cópia dos dados

```python
df1 = df.copy()
```

Remoção de colunas que não serão utilizadas

```python
df1 = df1.drop(["Unnamed: 0", "designation", "region_1", "region_2", "taster_name", "taster_twitter_handle"], axis=1)
df1.head()
```

```python
df1.info()
```

```python
print("{:.0%}".format(df1["price"].count() / df1.shape[0]))
```

Filtro somente das Colunas de País e Tipos de Uvas

```python
df1 = df1.dropna(subset=["country", "price", "variety"])
df1.info()
```

# Análise Exploratória

```python
plt.figure(figsize=(10, 7))
plt.hist(data=df1, x='points',
         bins=np.arange(80, df1['points'].max() + 1, 1),
         color='purple')

plt.xlim(80, 100)
plt.title('Distribuição dos pontos', fontsize=16)
plt.xlabel('Pontos')
plt.ylabel('Frequência')
```

TOP 10 - Paises

```python
country_top_10= df1['country'].value_counts()[:10]
country_top_10
```

Gráfico de países que mais tiveram vinhos avaliados

```python
plt.figure(figsize=(10, 7))
country_top_10.plot(kind='bar', color='purple')
plt.title('Número de classificações de vinhos por país', fontsize=16)
plt.xlabel('País')
plt.ylabel('Total de avaliações')
```

Países que produzem vinhos de melhor qualidade

```python
melhor_qualidade = df1.groupby('country')['points'].agg("mean")
melhor_qualidade = melhor_qualidade.sort_values(ascending=False)[:10]
melhor_qualidade.to_frame().style.background_gradient(cmap='Purples', high=0.5)
```

Geração de Gráfico a partir desses valores

```python
y = melhor_qualidade.index.to_series()
x = melhor_qualidade

plt.figure(figsize=(10, 7))
plt.xlim(80, 92)
sns.barplot(x=x, y=y, palette='Purples_r')
plt.title('Top 10 média de vinhos de melhor qualidade por país', fontsize=16)
plt.xlabel('Média de pontos')
plt.ylabel('País')
```

Países com pontuação acima de 95

```python
high_scores = df1[df1['points'] >= 95]
high_scores['country'].value_counts()
```

Gráfico de Maximo, minimo e média das pontuações por Países

```python
plt.figure(figsize=(20, 10))
sns.boxplot(data=df1, x='points', y='country',color='Purple')
plt.title('Pontuaçãopor país', fontsize=16)
plt.xlabel('Pontos')
plt.ylabel('País')
```

Portugal, EUA, Itália, França e Austrália recberam pontuação máxima

Preço dos Vinhos por Países

```python
average_price = df1.groupby('country')['price'].mean()
average_price = average_price.sort_values(ascending=False)
y= average_price.index.to_series()
x= average_price.values

plt.figure(figsize=(12, 8))
sns.barplot(x=x, y=y, palette='Purples_r')
plt.title('Preço médio da garrafa de vinhos por país', fontsize=16)
plt.xlabel('Preço médio ($)')
plt.ylabel('País')
```

Suíça produz o vinho mais caro de 80 dolares e Ucrania o mais barato, menor que 10 dolares


Verificar o preço por qualidade do vinho

```python
plt.figure(figsize=(10, 7))
bins = np.arange(0, df1["price"].max() + 5, 5)
plt.hist(data=df1, x='price', bins=bins, color='Purple')
plt.xlim(0, 200)
plt.title('Distribuição dos preços dos vinhos', fontsize=16)
plt.xlabel('Preço($)')
plt.ylabel('Frequência')
```

Transformação Logarítmica nos dados

```python
df1['price'].describe()
```

```python
#distribuição logarítmica
np.log10(df1['price']).describe()
```

Geração de Gráficos com a transformação logarítmica

```python
plt.figure(figsize=(12, 8))  # Define o tamanho da figura do gráfico
bins = 10 ** np.arange(0.5, 3.5 + 0.05, 0.05)  # Define os intervalos (bins) para o histograma usando escala logarítmica
ticks = [1, 3, 10, 30, 100, 300, 1000, 3000, 10000]  # Define os valores dos ticks (marcações) do eixo x
plt.hist(data=df1, x='price', bins=bins, color='Purple')  # Cria o histograma dos preços usando a cor roxa
plt.title('Distribuição dos preços dos vinhos', fontsize=16)  # Adiciona um título ao gráfico
plt.xscale('log')  # Define a escala do eixo x como logarítmica
plt.xticks(ticks, ticks)  # Define as marcações do eixo x como os valores de "ticks"
plt.xlabel('Preço ($)')  # Define o rótulo do eixo x
plt.ylabel('Frequência')  # Define o rótulo do eixo y
```

Correlação entre a pontuação e o valor da garrafa

```python
x = [1, 3, 10, 30, 100, 300, 1000, 3000, 10000]

plt.figure(figsize=(12, 8))
sns.regplot(x='price', y='points', data=df1, color='Purple', logx=True, scatter_kws={'alpha':0.3})
plt.title("Pontos vs. Preço", fontsize=14)
plt.xscale('log')
plt.xticks(x, x)
plt.xlabel("Preço($)")
plt.ylabel("Pontos")
```

Acima de 50

```python
#Top 10 até 15

under_15 = df1.drop(["description", "province", "winery"], axis=1)
under_15 = under_15[under_15["price"] <= 15]
under_15 = under_15.sort_values(by="points", ascending=False)[:10]
under_15.style.background_gradient(cmap='YlOrRd', high=0.5, subset=['price'])
```

```python
#Top 10 entre 15 a 30

under_15_and_30= df1.drop(["description", "province", "winery"], axis=1)
under_15_and_30 = under_15_and_30.query("price > 15 and price <= 30")
under_15_and_30 = under_15_and_30.sort_values(by="points", ascending=False)[:10]
under_15_and_30.style.background_gradient(cmap='YlOrRd', high=0.5, subset=['price'])
```

```python
#Top 10 entre 30 a 50

under_30_and_50= df1.drop(["description", "province", "winery"], axis=1)
under_30_and_50 = under_30_and_50.query("price > 30 and price <= 50")
under_30_and_50 = under_30_and_50.sort_values(by="points", ascending=False)[:10]
under_30_and_50.style.background_gradient(cmap='YlOrRd', high=0.5, subset=['price'])
```

```python
#Top 10 acima de 50

above_50= df1.drop(["description", "province", "winery"], axis=1)
above_50 = above_50.query("price > 50")
above_50 = above_50.sort_values(by="points", ascending=False)[:10]
above_50.style.background_gradient(cmap='YlOrRd', high=0.5, subset=['price'])
```

Palavras mais citadas nas avaliações dos vinhos

```python
from PIL import Image
from wordcloud import WordCloud, STOPWORDS

text = df1['description'][1]
text = " ".join(i for i in df1["description"])
len(text)
```

Retirada de Palavras que são repitidas e não trarão muitos benefícios para a análise

```python
stopwords = set(STOPWORDS)
stopwords.update(["drink", "now", "wine", "flavor", "flavors"])
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=2048, height=1024).generate(text)
```
